**The four adjacent digits in the 1000-digit number that have the greatest product are 9 × 9 × 8 × 9 = 5832.
Find the thirteen adjacent digits in the 1000-digit number that have the greatest product. What is the value of this product?**



```python
products = []
for n,v in enumerate(num_lst):
    four_list = []
    product = 1 
    for i in range(1,13):
        four_list.append(num_lst[n-i])
    four_list.append(v)
    for i in four_list:
        product *= i 
    products.append(product)
    #two_product = v *num_lst[n-1]
    #products.append(two_product)
print 'largest 13 digit product:', max(products)    
```
